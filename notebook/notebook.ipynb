```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
dirname = os.path.join(os.getcwd(), 'dataset')
datasets = {}
for file in os.listdir(dirname):
    car_type = file.split('.')[0]
    file_data = pd.DataFrame(pd.read_csv(os.path.join(dirname, file)))
    df = pd.concat([pd.DataFrame([{"type": car_type} for i in range(len(file_data))]), file_data], axis=1)
    print(file.split('.')[0])
    display(df)
    datasets[car_type] = df
```

```python
datasets['cclass']['type'] = 'merc'
datasets['focus']['type'] = 'ford'
display(datasets['focus'])
```
